```python
def clean_text(text):
    clean_pattern = r"[\[\]']"
    clean_text = re.sub(clean_pattern, '', text)
    return clean_text.strip()
```

```python
labels= [
    "Fiction",
    "Non-Fiction",
    "Mystery/Thriller",
    "Romance",
    "Science Fiction (Sci-Fi)",
    "Fantasy",
    "Horror",
    "Biography/Autobiography/Memoir",
    "Historical Fiction",
    "Young Adult (YA)",
    "Children's Books",
    "Self-Help",
    "Psychology",
    "Science/Nature",
    "History",
    "Travel",
    "Cookbooks/Cooking",
    "Religion/Spirituality",
    "Philosophy",
    "Poetry",
    "Art/Photography",
    "Business/Entrepreneurship",
    "Politics/Government",
    "Economics",
    "True Crime",
    "Health/Fitness",
    "Parenting/Family",
    "Education",
    "Science/Technology",
    "Sports",
    "DIY/Crafts",
    "Gardening",
    "LGBTQ+ Fiction/Non-Fiction",
    "Classic Literature",
    "Contemporary Literature",
    "Humor/Comedy",
    "Travelogue",
    "Essays",
    "Fantasy Anthologies",
    "Short Stories/Flash Fiction"
]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(categories_list + labels)
```

```python
cosine_similarities = cosine_similarity(tfidf_matrix[:-len(labels)], tfidf_matrix[-len(labels):])
# Find the most similar common book category for each category in the current category column
similar_categories_scores = cosine_similarities.max(axis=1)
similar_categories_indices = cosine_similarities.argmax(axis=1)
similar_common_book_categories = [labels[i] for i in similar_categories_indices]

# Replace the current categories with the most similar common book categories
df['new_categories'] = similar_common_book_categories

# Add a new column with the similarity score of the most similar category
df['similarity_score'] = similar_categories_scores
```

```python
from nltk.corpus import stopwords 
stopwords = list(set(stopwords.words('english')))
stopwords[:10]
```

```python
import string
punctuation = list(set(string.punctuation))
punctuation[:10]
```

```python
import re
from nltk.stem.wordnet import WordNetLemmatizer

def cleanup(text):
    '''
        input: raw text
        output: a list of words
    '''
    
    text = " ".join([i for i in text.lower().split() if i not in stopwords])           # Remove stopwords
    text = "".join(re.sub(r'[^a-zA-Z ]', '', i) for i in text if i not in punctuation) # Keep only alphabetical characters
    text = [WordNetLemmatizer().lemmatize(i) for i in text.split()]                    # Lemmatize
    return text
```

```python
from gensim.corpora.dictionary import Dictionary
```

```python
clean_titles = df['Title'].apply(cleanup)
clean_titles[:10]
```

```python
dictionary = Dictionary(clean_titles)
len(dictionary.items())
```

```python
dictionary.filter_extremes(no_below=40, no_above=0.7)

len(dictionary.items())
```

```python
corpora = [dictionary.doc2bow(doc) for doc in clean_titles]
```

```python
from gensim.models import ldamodel

#train an LDA model
lda_model = ldamodel.LdaModel(corpora, num_topics=30, id2word = dictionary, passes=120)
```

```python
topics = lda_model.get_document_topics(corpora)

clusters = [[i] for i in range(lda_model.num_topics)]
for doc_id, doc_topics in enumerate(topics):
    dominant_topic = max(doc_topics, key=lambda x: x[1])[0] 
    clusters[dominant_topic].append(doc_id)
```

```python
#output each cluster and top5 topics
for cluster_id, document_ids in enumerate(clusters):
    print(f"Cluster {cluster_id + 1}:")
    cluster_topics = lda_model.get_topic_terms(cluster_id, topn=6)
    for topic_id, _ in cluster_topics:
        topic_words = lda_model.id2word[topic_id]
        print(f"{topic_words}")
    print()
```

```python
#visualize the clusters and topics
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
pyLDAvis.enable_notebook()
vis = gensimvis.prepare(lda_model, corpora, dictionary)
vis
```

```python
from gensim.models import CoherenceModel
```

```python
# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=clean_titles, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('Coherence Score: ', coherence_lda)
```
